#### Plot Pie Chart

This functions plots a pie chart. The pie plot "explodes" all the slices.

```python
def plot_pie(labels: list[str], sizes: list[int], percentages: list[float], title: str):
    """
    Creates a pie chart.

    :param labels: Slice labels for the chart
    :param sizes: Slice sizes for the chart
    :param percentages: Percentages of sizes
    :param title: Title of the chart
    """
    fig, ax = plt.subplots()

    # Explode values
    explode = [0.05] * len(sizes)

    # Pie chart parameters
    patches = ax.pie(x=sizes, explode=explode, startangle=90, pctdistance=0.85)
    labels = [f"{label} ({perc:0.1f}%)" for label, perc in zip(labels, percentages)]
    ax.legend(patches[0], labels, loc="best")
    ax.set_title(title)

    # Creating center circle
    center_circle = Circle(xy=(0, 0), radius=0.7, fc="white")
    fig = plt.gcf()
    # Adding circle
    fig.gca().add_artist(center_circle)
    ax.axis("equal")

    plt
```

### Plots

```python
# List of unique donors
# --> Act as labels for the plot
labels = adata.obs["donor"].unique().tolist()

# Returns the amount of samples the condition fits
# --> Act as sizes for the plot
sizes = []
for donor in labels:
    sizes.append(adata[adata.obs["donor"] == donor].shape[0])

# Calculate percentages for visualization
total = adata.obs_names.shape[0]
percentages = [((size / total) * 100) for size in sizes]

plot_pie(
    labels=labels, 
    sizes=sizes,
    percentages=percentages,
    title="Sample Distribution per Donor"
)
```

```python
# List of unique cell types
# --> Act as labels for the plot
labels = adata.obs["cell_type"].unique().tolist()

# Returns the amount of samples the condition fits
# --> Act as sizes for the plot
sizes = []
for cell_type in labels:
    sizes.append(adata[adata.obs["cell_type"] == cell_type].shape[0])

# Calculate percentages for visualization
# Calculate percentages for visualization
total = adata.obs_names.shape[0]
percentages = [((size / total) * 100) for size in sizes]

plot_pie(
    labels=labels,
    sizes=sizes,
    percentages=percentages,
    title="Sample Distribution per Cell Type"
)
```

```python
def cpm_normalization(expression, total_expression):
    scaling_factor = 1e6
    cpm_normalized = (expression * scaling_factor) / total_expression
    
    return cpm_normalized
```

```python
def normalize_row_values(row, total_expression):
    for value in range(len(row)):
        row[value] = cpm_normalization(row[value], total_expression)
        
    return row
```

```python
def calculate_total_expression():
    total_expression = 0
    count = 0
    for row in adata.obs_names:
        row_values = adata[row].X.toarray()
        total_expression += row_values.sum()
        count += 1
        
        if (count % 10000) == 0:
            print(f"{count}/{len(adata.obs_names)}")
        
    return total_expression
```

```python
normalized_data = []
total_expression = calculate_total_expression()
print(f"Total expression calculated...{total_expression}")
count = 0
for row in adata.obs_names:    
    row_values = adata[row].X.toarray()
    normalized_row_values = normalize_row_values(row_values, total_expression)
    normalized_data.append(normalize_row_values)
    count += 1
    
    if (count % 10000) == 0:
        print(f"{count}/{len(adata.obs_names)}")
```
